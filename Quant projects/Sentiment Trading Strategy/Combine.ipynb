```python
import pandas as pd

# Specify the two CSV files (relative to /Algo/Thesis/)
file1 = 'S&P Updated/AMZN_updated.csv'  # Replace with your first CSV
file2 = 'S&P Updated/AMZN_2024_2025_wallstreetbets_updated.csv'  # Replace with your second CSV

try:
    # Load CSVs
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    
    # Combine CSVs
    combined_df = pd.concat([df1, df2], ignore_index=True)
    
    # Check for duplicates based on 'id'
    duplicate_count = combined_df.duplicated(subset='id').sum()
    if duplicate_count > 0:
        print(f"Warning: Found {duplicate_count} duplicate IDs. Removing duplicates...")
        combined_df = combined_df.drop_duplicates(subset='id', keep='first')
    
    # Ensure all expected columns
    expected_columns = [
        'type', 'id', 'parent_post_id', 'subreddit', 'title', 'text',
        'url', 'upvotes', 'created_utc', 'author',  'word_count', 'author_created_utc'
    ]
    for col in expected_columns:
        if col not in combined_df.columns:
            combined_df[col] = ''
    combined_df = combined_df[expected_columns]
    
    # Save combined CSV
    output_file = 'AMZN_updated.csv'
    combined_df.to_csv(output_file, index=False, encoding='utf-8')
    
    # Print summary
    print(f"Combined {file1} and {file2} into {output_file}")
    print(f"Rows in {file1}: {len(df1)}")
    print(f"Rows in {file2}: {len(df2)}")
    print(f"Total rows after combining: {len(combined_df)}")
    print(f"Columns: {list(combined_df.columns)}")
    
except Exception as e:
    print(f"Error combining files: {e}")
```
