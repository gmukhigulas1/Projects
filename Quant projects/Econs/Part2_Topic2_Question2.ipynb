```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv('Liabilities to Other Resident Sectors.csv').transpose().dropna()
```

```python
b = pd.read_csv('Transferable Deposits.csv').transpose().dropna()
```

```python
a= a[a[0] != 'M'].copy().iloc[1:]
b= b[b[0] != 'M'].copy().iloc[1:]
```

```python
a.columns = ['Liabilities to Other Depository Corporations']
b.columns = ['Transferable Deposits']
```

```python
model = pd.concat([a, b], axis=1)/1000000
```

```python
model.plot(figsize = (13,8), fontsize=12)
plt.title("Checkable deposits & Bank Reserves (Local Currency, in Millions)", fontsize = 20, color="Black")
plt.legend(fontsize=10)
plt.show()
```
