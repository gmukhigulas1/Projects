Lists are a collection of elements. You can create a list by using square brackets '[ ]'.

```python
list = [x, y, 4, 5]
```

Arrays can have multiple dimensions. You can create arrays with multiple dimensions by using the 'np.zeros()' function.

```python
array_2d = np.zeros((2, 3))
print('2D Array: ', array_2d)
```

You can access the elements of a 2D array by using two indices. The first index refers to the row, the second index refers to the column. You can also access the elements of a 2D array by using the ':' operator. The ':' operator refers to all elements of the respective axis.

```python
# change the first row of the 2D array
array_2d[0,:] = [1, 2, 3]
# change the second entry of the second row
array_2d[1,1] = 1
print('2D Array: ', array_2d)
# print the shape of the array
print('Shape: ', array_2d.shape)
print(array_2d)
```

You can transpose a 2D array by using the 'T' attribute. This flips the rows and columns of the array.

```python
print('Transpose: ', array_2d.T)
print('Shape: ', array_2d.T.shape)
array_2d
```

You can also perform more complex operations on arrays that are very useful for data analysis.

```python
print('Sum of all entries: ', np.sum(array_2d))
print('Mean of all entries: ', np.mean(array_2d))
print('Mean per column: ', np.mean(array_2d, axis=0))
print('Standard deviation among all entries: ', np.std(array_2d))
print('Variance among all entries: ', np.var(array_2d))
```

Two very useful operations when it comes to time series data are the cumulative sum and the difference of an array as they allow to switch between increments and the actual values. For $x\in\mathbb{R}^n$, the cumulative sum is defined as $y_i = \sum_{j=0}^i x_j$ and the difference is defined as $y_i = x_{i+1} - x_i$.

```python
print('Original array: ', array)
print('Cumulative sum: ', np.cumsum(array))
print('Recover increments: ', np.diff(np.cumsum(array)))
```

## Loops
Loops are used to execute a block of code multiple times. The most common loop is the 'for' loop. Note that indentation is crucial in Python. The block of code that belongs to the loop is indented.

```python
# create an array with n=10 0-entries
n = 10
array = np.zeros(n)

# range(n) creates a list of integers from 0 to n-1 which is the iterable the index i will run over
for i in range(n):
    # this block of code is executed n times with i being the current index
    print('Current index: ', i)
```

## Functions
Functions are used to create reusable code blocks that in the end return a certain result. You define a function by using the 'def' keyword followed by the name of the function and the arguments in parentheses. The block of code that belongs to the function is indented (this is again crucial!). The 'return' statement in the end defines the value the function returns to the user.

```python
# here we define a function that manipulates the input x
def manipulate(x):
    # if statements are executed only when the corresponding condition evaluates to 'True'; otherwise the else statement is executed
    if x < 2:
        y = x ** 2
    else:
        y = 4 - (x - 2) ** 2
    return y
```

 After defining it, you can call a function by using its name followed by the arguments in parentheses.

```python
print('Call the function Manipulate on the input 1.5: ', manipulate(1.5))
```

You can create plots by using the 'matplotlib' package. The most common plot for functions and paths of random processes is the line plot. You can create a line plot by using the 'plt.plot()' function. To show the plot, you have to use the 'plt.show()' function.

## Random Number Generation
Simulation of random numbers can be done by the 'numpy' package. However, computers can't produce truly random numbers, instead they generate pseudo-random numbers which only appear to be random. By setting the seed of the random number generator, you can make the random number in your code reproducable - executing the code at a later point in time will produce exactly the same random numbers and makes your code comparable. The seed is set by using the 'np.random.seed()' function.

The most common distribution is the normal distribution, which takes as arguments the mean, standard deviation and number of points that should be sampled.

```python
# make the random numbers reproducible by setting the seed - here we set the seed to 213, but it could be any integer
np.random.seed(213)
# sample 1000 random numbers from a normal distribution with mean 3 and standard deviation 2 and save them in the array 'data'
data = np.random.normal(3.0, 2.0, 1000)
```

To understand the distribution of the data, you can plot a histogram. You can create a histogram by using the 'plt.hist()' function. The 'bins' argument specifies the number of intervals used to create the histogram. The 'density' argument normalizes the histogram such that the area under the histogram is equal to 1.

```python
# the scipy.stats package contains density functions for different distributions and statistical functions
import scipy.stats as scs

# use the histogram function contained in matplotlib to plot the histogram of the data
plt.hist(data, bins=20, density=True)
# plot theoretical normal distribution as comparison
x = np.linspace(-3, 12, 100)
plt.plot(x, scs.norm.pdf(x, 3.0, 2.0))

plt.show()
```

Statistics of data can be computed by using the 'numpy' package as before. You can compute the mean, standard deviation, skewness, kurtosis, and the percentage of positive values.

```python
# analyze the data
print('Amount of data points: ', len(data))
print('Mean: ', np.mean(data))
print('Standard deviation: ', np.std(data))
print('Skewness: ', scs.skew(data))
print('Kurtosis: ', scs.kurtosis(data))
print('Percentage of positive values: ', np.sum(data > 0) / len(data) * 100, '%')
```

```python
# use pandas to summarize the statistics
import pandas as pd
```

```python
# condense statistics into a pandas dataframe
stats = pd.DataFrame(data=[len(data), np.mean(data), np.std(data), scs.skew(data), scs.kurtosis(data), np.sum(data > 0) / len(data) * 100],
                     index=['Number of points', 'Mean', 'Standard deviation', 'Skewness', 'Kurtosis', 'Percentage of positive values'],
                     columns=['Statistics'])
print(stats)
```

## Scatter Plots
Scatter plots are useful to visualize the relationship between two variables. You can create a scatter plot by using the 'plt.scatter()' function. A scatter plot usually takes two data sets as input and plots the data points in a two-dimensional space.

## QQ-Plots
QQ-Plots are used to compare the distribution of a sample with a theoretical distribution, by default the standard normal distribution. If the sample data is normally distributed, the QQ-Plot should be a straight line, if it is standard normally distributed, the QQ-Plot should be the line $y = x$.

```python
import statsmodels.api as sm
```

```python
sm.qqplot(data, line='45')
plt.show()

# you can see the data seems to be normally distributed but not standard normally distributed
```

```python
# normalize the data to make it standard normally distributed
sm.qqplot((data - 3.0) / 2.0, line='45')
plt.show()
# now the data is standard normally distributed
```

#### ToDo: Define a function 'simulate' that automates the simulation of one trajectory of the process with n steps.
Hint: The function has one argument n and returns one trajectory of the process with n+1 values.

```python
def simulate(n):
    x= np.random.choice([-1,1], n+1)
    x[0]=0
    
    return np.cumsum(x)
```

```python
stockprice = np.zeros((2,2))
P0=100
u=1.17
d=0.86
stockprice[0,0]=P0
stockprice[1,0]=P0*u
stockprice[1,1]=P0*d
stockprice[0,1]=0
print(stockprice)
```

```python
def stockprice(P0, u, d):
    stockprice=np.zeros((2,2))
    stockprice[0,0]=P0
    stockprice[1,0]=P0*u
    stockprice[1,1]=P0*d
    stockprice[0,1]=0
    return stockprice
```

```python
def callpayoff(P , K):
    array = np.empty(2)
    array[0]=max(P[1,0]-K,0)
    array[1]=max(P[1,1]-K,0)
    return array
```

```python
P = stockprice(100,1.17, 0.86)
print(callpayoff(P,101))
```

```python
r=0.01
impVol=0.2
```

```python
def fitU(r, impVol):
    vol = impVol**2
    e = np.exp(vol)
    return (1+(1+r)**2*e+np.sqrt((1+(1+r)**2*e)**2-4*(1+r)**2))/(2*(1+r))
```

```python
u = fitU(r,impVol)
d= 1/u
```

```python
callpayoff(stockprice(100,fitU(r,impVol), 1/fitU(r,impVol)),80)
```

```python
def callprice(P0, u, d,r,K):
    q = (1+r-d)/(u-d)
    D1= callpayoff(stockprice(P0,u, d),K)
    D0 = (D1[0]*q+D1[1]*(1-q))/(1+r)
    return D0
```

```python
callprice(100,u,d,r,80)
```

```python
from math import comb
def price(P0, K, r, sigma, T):
    u= fitU(r,sigma)
    d=1/u
    
    arr = np.zeros((T+1, T+1))
    for i in range(T+1):
        for j in range(T+1):
            arr[i, j] = P0 * (u ** j) * (d ** (i - j))
            
    DT = np.empty(T+1)
    for i in range(T+1):
        DT[i]=max(arr[T,i]-K,0)

    q = (1+r-d)/(u-d)

    price = 0.0
    for i in range(T+1):
        prob = comb(T, i) * q**i * (1-q)**(T-i)
        price += prob * DT[i]

    price = price / (1+r)**T
    return price
```

```python
print(price(P0=100, K=101, r=0.02, sigma=0.15, T=2))
```
